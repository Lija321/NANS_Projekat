```python
df_comaprison=pd.read_csv('./euroleague_simple/euroleague_comparison.csv')
df_advanced=pd.read_csv('./euroleague_advanced/games_advanced_stats.csv')
```

```python
df_comaprison['assists_a'] = df_comaprison['assists_bench_a'] + df_comaprison['assists_starters_a']
df_comaprison['assists_b'] = df_comaprison['assists_bench_b'] + df_comaprison['assists_starters_b']
df_comaprison['turnovers_a'] = df_comaprison['turnovers_bench_a'] + df_comaprison['turnovers_starters_a']
df_comaprison['turnovers_b'] = df_comaprison['turnovers_bench_b'] + df_comaprison['turnovers_starters_b']
df_comaprison['steals_a'] = df_comaprison['steals_bench_a'] + df_comaprison['steals_starters_a']
df_comaprison['steals_b'] = df_comaprison['steals_bench_b'] + df_comaprison['steals_starters_b']
df_comaprison['poinst_a'] = df_comaprison['points_bench_a'] + df_comaprison['points_starters_a']
df_comaprison['poinst_b'] = df_comaprison['points_bench_b'] + df_comaprison['points_starters_b']

columns_to_drop=[ 
    "assists_bench_a",
    "assists_starters_a",
    "assists_bench_b",
    "assists_starters_b",
    "turnovers_bench_a",
    "turnovers_starters_a",
    "turnovers_bench_b",
    "turnovers_starters_b",
    "steals_bench_a",
    "steals_starters_a",
    "steals_bench_b",
    "steals_starters_b",
    "points_bench_a",
    "points_starters_a",
    "points_bench_b",
    "points_starters_b",
    ]
df_comaprison.drop(columns_to_drop, inplace=True, axis=1)
df_comaprison=df_comaprison[df_comaprison['phase']=="REGULAR SEASON"]
df_comaprison=df_comaprison[df_comaprison['season_code']>='E2016']
df_comaprison=df_comaprison[df_comaprison['season_code']<='E2017']
```

```python
count=0
for row in df_comaprison.iterrows():
    row=row[1]
    if row['round']==1: #Skip first round because we don't have data for previous round
        continue

    team_a=row['team_id_a']
    #print(f"Team a: {team_a}\n====================")
    team_b=row['team_id_b']
   #print(f"Team b: {team_b}\n====================")

    game_id=row['game_id']
    #print(f"Game id: {game_id}\n====================")

    previous_game_a=df_comaprison[(df_comaprison['season_code']==row['season_code']) & (df_comaprison['round']==row['round']-1) & ((df_comaprison['team_id_a']==team_a) | (df_comaprison['team_id_b']==team_a))]
    previous_game_b=df_comaprison[(df_comaprison['season_code']==row['season_code']) & (df_comaprison['round']==row['round']-1) & ((df_comaprison['team_id_a']==team_b) | (df_comaprison['team_id_b']==team_b))]

    #print(f"Previous game a:\n{previous_game_a.iloc[0]}\n====================")
    #print(f"Previous game b:\n{previous_game_b.iloc[0]}\n====================")
    
    try:
        previous_game_a=previous_game_a.iloc[0]
        previous_game_b=previous_game_b.iloc[0]
    except:
        print(f"Error in game {game_id}")
        print(f"{df_comaprison[df_comaprison['game_id']==game_id]}")
        print(f"Count: {count}")
        print(previous_game_a)
        break


    #Team a
    if len(previous_game_a)!=0:
        if previous_game_a['team_id_a']==team_a:
            A_game_id=game_id+"_a"
            A_team_id=team_a
            A_opponent=team_b
            A_opponent_previous_game=previous_game_a['team_id_b']
            A_points_socred=previous_game_a['poinst_a']
            A_fast_break_points_for=previous_game_a['fast_break_points_a']
            A_fast_break_points_against=previous_game_a['fast_break_points_b']
            A_turnover_points_for=previous_game_a['turnover_points_a']
            A_turnover_points_against=previous_game_a['turnover_points_b']
            A_second_chance_points_for=previous_game_a['second_chance_points_a']
            A_second_chance_points_against=previous_game_a['second_chance_points_b']
            A_defensive_rebounds_for=previous_game_a['defensive_rebounds_a']
            A_defensive_rebounds_against=previous_game_a['defensive_rebounds_b']
            A_offensive_rebounds_for=previous_game_a['offensive_rebounds_a']
            A_offensive_rebounds_against=previous_game_a['offensive_rebounds_b']
            A_turnovers_for=previous_game_a['turnovers_a']
            A_turnovers_against=previous_game_a['turnovers_b']
            A_steals_for=previous_game_a['steals_a']
            A_steals_against=previous_game_a['steals_b']
            A_assists_for=previous_game_a['assists_a']
            A_assists_against=previous_game_a['assists_b']

        else:
            A_game_id=game_id+"_a"
            A_team_id=team_a
            A_opponent=team_b
            A_opponent_previous_game=previous_game_a['team_id_a']
            A_points_socred=previous_game_a['poinst_b']
            A_fast_break_points_for=previous_game_a['fast_break_points_b']
            A_fast_break_points_against=previous_game_a['fast_break_points_a']
            A_turnover_points_for=previous_game_a['turnover_points_b']
            A_turnover_points_against=previous_game_a['turnover_points_a']
            A_second_chance_points_for=previous_game_a['second_chance_points_b']
            A_second_chance_points_against=previous_game_a['second_chance_points_a']
            A_defensive_rebounds_for=previous_game_a['defensive_rebounds_b']
            A_defensive_rebounds_against=previous_game_a['defensive_rebounds_a']
            A_offensive_rebounds_for=previous_game_a['offensive_rebounds_b']
            A_offensive_rebounds_against=previous_game_a['offensive_rebounds_a']
            A_turnovers_for=previous_game_a['turnovers_b']
            A_turnovers_against=previous_game_a['turnovers_a']
            A_steals_for=previous_game_a['steals_b']
            A_steals_against=previous_game_a['steals_a']
            A_assists_for=previous_game_a['assists_b']
            A_assists_against=previous_game_a['assists_a']

        df_parsed.loc[count]=[ A_game_id
                              ,A_team_id
                              ,A_opponent
                              ,A_opponent_previous_game
                              ,A_points_socred
                              ,A_fast_break_points_for
                              ,A_fast_break_points_against
                              ,A_turnover_points_for
                              ,A_turnover_points_against
                              ,A_second_chance_points_for
                              ,A_second_chance_points_against
                              ,A_defensive_rebounds_for
                              ,A_defensive_rebounds_against
                              ,A_offensive_rebounds_for
                              ,A_offensive_rebounds_against
                              ,A_turnovers_for
                              ,A_turnovers_against
                              ,A_steals_for
                              ,A_steals_against
                              ,A_assists_for
                              ,A_assists_against
                              ,float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan')]
        count+=1



    #Team b
    if len(previous_game_b)!=0:
        if previous_game_b['team_id_a']==team_b:
            B_game_id=game_id+"_b"
            B_team_id=team_b
            B_opponent=team_a
            B_opponent_previous_game=previous_game_b['team_id_b']
            B_points_socred=previous_game_b['poinst_a']
            B_fast_break_points_for=previous_game_b['fast_break_points_a']
            B_fast_break_points_against=previous_game_b['fast_break_points_b']
            B_turnover_points_for=previous_game_b['turnover_points_a']
            B_turnover_points_against=previous_game_b['turnover_points_b']
            B_second_chance_points_for=previous_game_b['second_chance_points_a']
            B_second_chance_points_against=previous_game_b['second_chance_points_b']
            B_defensive_rebounds_for=previous_game_b['defensive_rebounds_a']
            B_defensive_rebounds_against=previous_game_b['defensive_rebounds_b']
            B_offensive_rebounds_for=previous_game_b['offensive_rebounds_a']
            B_offensive_rebounds_against=previous_game_b['offensive_rebounds_b']
            B_turnovers_for=previous_game_b['turnovers_a']
            B_turnovers_against=previous_game_b['turnovers_b']
            B_steals_for=previous_game_b['steals_a']
            B_steals_against=previous_game_b['steals_b']
            B_assists_for=previous_game_b['assists_a']
            B_assists_against=previous_game_b['assists_b']
            #TODO advanced stats
        else:
            B_game_id=game_id+"_b"
            B_team_id=team_b
            B_opponent=team_a
            B_opponent_previous_game=previous_game_b['team_id_a']
            B_points_socred=previous_game_b['poinst_b']
            B_fast_break_points_for=previous_game_b['fast_break_points_b']
            B_fast_break_points_against=previous_game_b['fast_break_points_a']
            B_turnover_points_for=previous_game_b['turnover_points_b']
            B_turnover_points_against=previous_game_b['turnover_points_a']
            B_second_chance_points_for=previous_game_b['second_chance_points_b']
            B_second_chance_points_against=previous_game_b['second_chance_points_a']
            B_defensive_rebounds_for=previous_game_b['defensive_rebounds_b']
            B_defensive_rebounds_against=previous_game_b['defensive_rebounds_a']
            B_offensive_rebounds_for=previous_game_b['offensive_rebounds_b']
            B_offensive_rebounds_against=previous_game_b['offensive_rebounds_a']
            B_turnovers_for=previous_game_b['turnovers_b']
            B_turnovers_against=previous_game_b['turnovers_a']
            B_steals_for=previous_game_b['steals_b']
            B_steals_against=previous_game_b['steals_a']
            B_assists_for=previous_game_b['assists_b']
            B_assists_against=previous_game_b['assists_a']
        
        df_parsed.loc[count]=[ B_game_id
                              ,B_team_id
                              ,B_opponent
                              ,B_opponent_previous_game
                              ,B_points_socred
                              ,B_fast_break_points_for
                              ,B_fast_break_points_against
                              ,B_turnover_points_for
                              ,B_turnover_points_against
                              ,B_second_chance_points_for
                              ,B_second_chance_points_against
                              ,B_defensive_rebounds_for
                              ,B_defensive_rebounds_against
                              ,B_offensive_rebounds_for
                              ,B_offensive_rebounds_against
                              ,B_turnovers_for
                              ,B_turnovers_against
                              ,B_steals_for
                              ,B_steals_against
                              ,B_assists_for
                              ,B_assists_against,
                              float('nan'),
                              float('nan'),
                              float('nan'),
                              float('nan'),
                              float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan'),float('nan')]
        count+=1


    df_parsed.loc[count-1,'opponent_steals_against']=A_steals_against
    df_parsed.loc[count-1,'opponent_steals_for']=A_steals_for
    df_parsed.loc[count-1,'opponent_defensive_rebounds_against']=A_defensive_rebounds_against
    df_parsed.loc[count-1,'opponent_defensive_rebounds_for']=A_defensive_rebounds_for
    df_parsed.loc[count-1,'opponent_offensive_rebounds_against']=A_offensive_rebounds_against
    df_parsed.loc[count-1,'opponent_offensive_rebounds_for']=A_offensive_rebounds_for
    df_parsed.loc[count-1,'opponent_turnovers_against']=A_turnovers_against
    df_parsed.loc[count-1,'opponent_turnovers_for']=A_turnovers_for
    
    df_parsed.loc[count-2,'opponent_steals_against']=B_steals_against
    df_parsed.loc[count-2,'opponent_steals_for']=B_steals_for
    df_parsed.loc[count-2,'opponent_defensive_rebounds_against']=B_defensive_rebounds_against
    df_parsed.loc[count-2,'opponent_defensive_rebounds_for']=B_defensive_rebounds_for
    df_parsed.loc[count-2,'opponent_offensive_rebounds_against']=B_offensive_rebounds_against
    df_parsed.loc[count-2,'opponent_offensive_rebounds_for']=B_offensive_rebounds_for
    df_parsed.loc[count-2,'opponent_turnovers_against']=B_turnovers_against
    df_parsed.loc[count-2,'opponent_turnovers_for']=B_turnovers_for


    
```

```python
for i,row in enumerate(df_parsed.iterrows()):
    data=row[1]
    game_id=data['game_id']
    season_code=data['game_id'][1:5]
    advanced_stats_games=df_advanced[(df_advanced['season']==int(season_code)) & (df_advanced['team']==data['team_id']) & (df_advanced['OP_team']==data['opponent'])]
    advanced_stats_games_opponent=df_advanced[(df_advanced['season']==int(season_code)) & (df_advanced['team']==data['opponent']) & (df_advanced['OP_team']==data['opponent_previous_game'])]
    if len(advanced_stats_games)!=0:
        df_parsed.loc[i,'3P%_for']=advanced_stats_games['3P%'].values[0]
        df_parsed.loc[i,'3P%_against']=advanced_stats_games['OP_3P%'].values[0]
        df_parsed.loc[i,'2P%_for']=advanced_stats_games['2P%'].values[0]
        df_parsed.loc[i,'2P%_against']=advanced_stats_games['OP_2P%'].values[0]
        df_parsed.loc[i,'FT%_for']=advanced_stats_games['FT%'].values[0]
        df_parsed.loc[i,'PFR_for']=advanced_stats_games['PFR'].values[0]
        df_parsed.loc[i,'PFD_against']=advanced_stats_games['PFDR'].values[0]
        df_parsed.loc[i,'3PR_for']=advanced_stats_games['3PR'].values[0]
        df_parsed.loc[i,'3PR_against']=advanced_stats_games['OP_3PR'].values[0]
        df_parsed.loc[i,'FTR_for']=advanced_stats_games['FTR'].values[0]
        df_parsed.loc[i,'FTR_against']=advanced_stats_games['OP_FTR'].values[0]
        df_parsed.loc[i,'PACE']=advanced_stats_games['PACE'].values[0]
        df_parsed.loc[i,'ORtg']=advanced_stats_games['ORtg'].values[0]
        df_parsed.loc[i,'DRtg']=advanced_stats_games['DRtg'].values[0]
        df_parsed.loc[i,'eFG%']=advanced_stats_games['eFG%'].values[0]
        df_parsed.loc[i,'TS%']=advanced_stats_games['TS%'].values[0]
    if len(advanced_stats_games_opponent)!=0:
        df_parsed.loc[i,'opponent_PFD_for']=advanced_stats_games_opponent['PFDR'].values[0]   
        df_parsed.loc[i,'opponent_PFD_against']=advanced_stats_games_opponent['PFR'].values[0]
        df_parsed.loc[i,'opponent_PACE']=advanced_stats_games_opponent['PACE'].values[0]
        df_parsed.loc[i,'opponent_ORtg']=advanced_stats_games_opponent['ORtg'].values[0]
        df_parsed.loc[i,'opponent_DRtg']=advanced_stats_games_opponent['DRtg'].values[0]

```
